```python
#imports
import pandas as pd
import numpy as np
```

# a. Understanding the dataset

```python
# 1. load the data
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
# 2. Show the number of attributes (columns) and number of records (rows)
print("Number of attributes (columns):", len(df.columns))
print("Number of records (rows):", len(df))
print()

# 3. Show the statistics of the dataset ( column wise mean, standard deviation, max, min, etc)
df.describe()
```

```python
# 5. Count the number of duplicate values in the dataset
print("duplicate values in the dataset:", df.duplicated().sum())

df[df.duplicated(keep=False)]
```

# b. Data cleaning

```python
# show the data types of the columns
df.dtypes
```

```python
# 1. Replacing the missing values with the mean of the column(if any)

# 1.1. Replacing the missing numerical values with the mean of the column
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# 1.2. Replacing the missing categorical values with the mode of the column
# categorical_cols = df.select_dtypes(include=[np.object]).columns
# df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
# 2. Drop the duplicate values from the dataset
df.drop_duplicates(inplace=True)
df
```

```python
# 3. If any row in the target column (Attrition) is missing, drop that row
df.dropna(subset=['Attrition'],inplace=True)
df
```

# c. Creation of input and output features

```python
# 1. Split the dataset into 2 parts: Features and Target
Features = df.drop(columns=['Attrition'], axis=1)
Labels = df['Attrition']
```

# d. Conversion of features into numeric values

```python
from sklearn.preprocessing import LabelEncoder

# convert the target column to numerical
le = LabelEncoder()
Labels = le.fit_transform(Labels)

Labels
# print(dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
# Convert the non-numeric columns into categorical columns
print(Features.dtypes)

categorical_cols = Features.select_dtypes(include=['object']).columns
Features[categorical_cols] = Features[categorical_cols].astype('category')

# 2. Convert the categorical columns into numeric columns
Features = pd.get_dummies(Features)

# Features.dtypes
# Features
```

# e. Scaling Features

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scaling(Features, scaler):

    one_hot_encoded_features = [col for col in Features.columns if set(Features[col].unique()) == {0, 1}]
    
    numeric_cols = Features.columns.difference(one_hot_encoded_features)

    scaled_features = Features.copy()
    scaled_features[numeric_cols] = scaler.fit_transform(Features[numeric_cols])
    
    return scaled_features
```

```python
scaler = StandardScaler()
# scaler = MinMaxScaler()

scaled_features = scaling(Features,scaler)

scaled_features
```

# f. Correlation Analysis

```python
features_df = pd.DataFrame(scaled_features, columns=Features.columns) # scaled feature data frame
target_df = pd.DataFrame(Labels, columns=['Attrition']) # target data frame

features_df
```

```python
# correlation analysis of features with target
target_series = target_df['Attrition']
correlations = features_df.corrwith(target_series)
correlations
```

```python
# top 20 features name with highest correlation with target
top_20_features = correlations.abs().sort_values(ascending=False).head(20)
# top_20_features

top_20_features_names = top_20_features.index
top_20_features_names
```

# 1D Scatter Plotter

```python
import matplotlib.pyplot as plt

plot = False
if plot:
    for feature in top_20_features_names:
        # separate the data based on numeric labels
        class_0 = features_df.loc[target_df['Attrition'] == 0]
        class_1 = features_df.loc[target_df['Attrition'] == 1]

        # plt.figure(figsize=(10, 6))
        plt.plot(class_0[feature], np.zeros_like(class_0[feature]), 'o', label='Class 0')
        plt.plot(class_1[feature], np.zeros_like(class_1[feature]), 'o', label='Class 1')

        plt.legend()
        plt.xlabel(feature)
        plt.title(f"1D Scatter Plot of {feature} by Attrition Classes")
        plt.show()
        plt.close()
```

# g. Validating the Pipeline

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(features_df, target_df, test_size=0.2, random_state=42)

# initialize the model
clf = LogisticRegression()

# train the model
clf.fit(X_train, y_train)

# predict the target values
y_pred = clf.predict(X_test)

# evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of Logistic Regression classifier: {accuracy:.2f}")
```
